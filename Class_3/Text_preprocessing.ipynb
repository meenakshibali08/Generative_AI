###Handle Punctuation -

String Library - Has lot of Punctuations

```python
import string
```

```python
string.punctuation
```

```python
exclude = string.punctuation
```

```python
for i in exclude:
    print(i)
```

```python
def remove_punct(text):
    for i in exclude:
        text = text.replace(i, "")
    return text
```

```python
text ="String @ with punctuation!!@ checking"
```

```python
remove_punct(text)
```

```python
text1 = "FYI this is not true"
text2 ="LAMO what is this"
text3 ="Please do it as ASAP!!"
```

```python
chat_words = {
"FYI" :"For your information",
"ASAP":"As soon as possible!!"}


```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)

    return " ".join(new_text)
```

```python
chat_conversion(text1)
```

```python
nltk.download("stopwords")
 ### stopwords is a collection of commonly used words in a language that are often removed from text during preprocessing 
 # because they do not carry significant meaning. Examples include "the", "is", "in", etc.
```

```python
stopwords.words("english")
```

```python
def remove_stopwords(sent):
    new_text =[]
    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())
        return " ".join(new_text)        
```

```python
sent="Hi my name is Meenakshi and I am learning Nlp"
```

```python
remove_stopwords(sent)
```

##Stemming - Converting the word into root form(not Accurate)
### Lemmitization - Bare form/Original form( Accurate)

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj = PorterStemmer()
    stem_words = [obj.stem(word) for word in text.split()]
    return stem_words
```

```python
input  = "The quick brown foxes jumping over the lazi dog"
```

```python
stemming(input)
```

```python
input_1 ="The striped bats were hanging on their feet and ate best fishes."
```

```python
from nltk.stem import WordNetLemmatizer

def lemmitization(text):
    obj = WordNetLemmatizer()
    lemma_words = [obj.lemmatize(word)for word in text.split()]
    return lemma_words
```

```python
lemmitization(input_1)
```
